```python
from mlxtend.frequent_patterns import association_rules,apriori
from mlxtend.preprocessing import TransactionEncoder
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
transactions = data.iloc[:, 0].str.split(',')
transactions
```

```python
te=TransactionEncoder()
te_data=te.fit(transactions).transform(transactions)
df_encoded=pd.DataFrame(te_data,columns=te.columns_)
df_encoded
```

### Apriori Algorithm

```python
frequent_itemsets=apriori(df_encoded,min_support=0.01,use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.tail(15)
```

# Interview Questions

### 1.	What is lift and why is it important in Association rules?
#### sol: Lift is a measure which is used to identify the relation between the items in the dataset. It evalutes the strength of an item that how much likely that consequent will occur for the given anticedent(Condition)
1. if lift>1 strong relation
2. if lift=1 both are independent
3. if lift<1 negative relation
#### The formula for calculating lift is:
###### Lift(𝐴→𝐵)=𝑃(𝐴∩𝐵)/𝑃(𝐴)⋅𝑃(𝐵)
​#Where:
P(A∩B) is the probability that both A and B occur together.
P(A) is the probability of A occurring.
P(B) is the probability of B occurring.

### 2.	What is support and Confidence. How do you calculate them?
#### Support helps identify frequently occurring items or patterns in the data, which is useful for identifying relevant itemsets.
Support(A)= Number of transactions containing A / Total number of transactions
#For a rule
 A→B, the support is calculated as:
Suppor(𝐴→𝐵)=Number of transactions containing both 𝐴 and 𝐵 / Total number of transactions
#### Confidence evaluates the predictive strength of a rule, indicating the reliability of using one item to predict another.
Confidence(A→B)=Support(A∩B) / Support(A)
#where:
A∩B represents the transactions that contain both A and B.
Support(A) is the frequency of A in the dataset.

# 3.	What are some limitations or challenges of Association rules mining?
1. High Computational And Complexity (Especially for large data sets requires more memory for huge number of items)
2. It has more number of rules which are not usefull everytime
3. Association rule mining is typically suited for categorical data and struggles with continuous data.
4. Rules that contain rare but valuable items may be missed because they don’t meet minimum support thresholds.
