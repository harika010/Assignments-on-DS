PCA(Principle Component Analysis

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('wine.csv')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
for cols in data:
    plt.figure(figsize=(5,10))
    plt.hist(data[cols])
    plt.xlabel='cols'
    plt.ylabel='values'
```

```python
for cols in data:
    plt.figure(figsize=(5,5))
    plt.scatter(x=data[cols],y=data[cols])
```

```python
data.corr()
```

```python
import seaborn as sn
```

```python
sn.heatmap(data.corr(),cmap='coolwarm')
```

```python
from sklearn.preprocessing import scale
```

```python
new_data=scale(data)
new_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=14)
```

```python
pca_values=pca.fit_transform(new_data)
```

```python
pca_values
```

```python
pca.components_ #calculates weight
```

```python
var=pca.explained_variance_ratio_
```

```python
var1=np.cumsum(np.round(var,4))*100
```

```python
var1
```

```python
plt.plot(var1,color='red')
```

```python
pca_values[:,0:6]
```

```python
final_data=pd.concat([pd.DataFrame(pca_values[:,0:6],columns=['pca1','pca2','pca3','pca4','pca5','pca6']),data['Alcohol']],axis=1)
```

```python
final_data
```

```python
sn.scatterplot(data=final_data,x='pca4',y='pca5',hue='Alcohol')
```

### Clustering With Original Data

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=5).fit(new_data)
```

```python
kmeans
```

```python
WCSS=[]
for i in range(1,14):
    clf=KMeans(n_clusters=i).fit(new_data)
    WCSS.append(clf.inertia_)
plt.plot(range(1,14),WCSS)
```

```python
clf=KMeans(n_clusters=2).fit(new_data)
```

```python
y_kmeans=clf.predict(new_data)
```

```python
y_kmeans
```

```python
labels=pd.Series(y_kmeans)
data['Clust']=labels
data
```

```python
data.groupby(data.Clust).sum()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(new_data[:,2],new_data[:,3],c=y_kmeans,cmap='coolwarm')
plt.scatter(kmeans.cluster_centers_[:,2],kmeans.cluster_centers_[:,3],c='red',alpha=0.75,label='Centriods')
```

```python
from sklearn.metrics import silhouette_score
```

```python
score=silhouette_score(data,y_kmeans)
```

```python
score
```

```python
WCSS
```

### Clustering With PCA data

```python
kmeans=KMeans(n_clusters=2).fit(final_data)
```

```python
y_kmeans=kmeans.predict(final_data)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
WCSS=[]
for i in range(1,14):
    clf=KMeans(n_clusters=i).fit(final_data)
    WCSS.append(clf.inertia_)
plt.plot(range(1,14),WCSS)
```

```python
score=silhouette_score(final_data,kmeans.labels_)
```

```python
score
```

```python
final_data
```

```python
plt.scatter(data=final_data,x='pca2',y='pca3',c=y_kmeans,cmap='coolwarm')
```

###### There is no much difference on the Original Data and PCA (reduced dimensions )Data So we can predict the model performance by Reducing the No of Dimensions and understanding the patterns more eaisly

Conclusion:
The patterns for more no of dimensions is not clear so use pca to reduce the no of dimensions
While Reducing the no of dimensions using PCA will not effect the performance of the model 
We may increase the performance of the model with less no of dimensions and capture maximum information
