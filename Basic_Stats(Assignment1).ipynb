```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Ploting Histograms for the numerical_columns
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.grid(True)
    plt.show()
```

```python
#Finding kurtosis(Highest Peak)
numerical_columns.kurt()
#Volume have high kurt value so it has more outliers
#The kurtosis value less than 3 is left skewed
```

```python
#Finding Standard Deviation for the numerical_columns
def Standardized_column(column):
  mean=column.mean()
  std=column.std()
  Standardized_column=(column-mean)/std
  return Standardized_column
```

```python
#Applying Standardization for each columns
for column in numerical_columns:
  df[column]=Standardized_column(df[column])
#Now numerical columns are standardized
print(df[numerical_columns.columns].head())
```

```python
#Creating the Dummy values for numerical_columns
df_dummy=pd.get_dummies(df[numerical_columns.columns])
df_dummy=df_dummy.astype(int)
```
