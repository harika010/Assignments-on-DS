```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoding
```

### One-Hot Encoding
#### Pros:
1. No Assumptions for Nominal Data(categorical_data without a natural order)
2. Effective for Non linear data such as Decision Tress,Random Forest,Neural Networks
3. Each category is clearly seperated making it easy to understand each individual features eaisly
#### Cons:
1. Increased Dimesionality which may increase computational cost and memory usage
2. Less Effective for Ordinal Data
### Label Encoding
#### Pros:
1. Simplycity: It doesn't increase the no of dimensions
2. More Efficient Making it suitable for the datasets with large no of categorical features
3. It is suitable for ordinal data
#### Cons:
1. It implies ordinal relationship for nominal data as well
2. Always it will not effective for Non linear Data

### 3. Feature Engineering

```python
# Check for skewness in numerical features
skewness = df1[['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']].skew()
skewness
# capital_gain and capital_loss are more possitively skewed
```

```python
numerical_cols
```

```python
bins=[0,30,45,60,100]
labels=['18-30','31-45','46-60','60+']
df1['age_group']=pd.cut(df1['age'],bins=bins,labels=labels,right=False)
```

```python
df1['capital_gain_log']=np.log1p(df1['capital_gain'])
df1['capital_loss_log']=np.log1p(df1['capital_loss'])
```

```python
df1['income_per_hour']=df1['capital_gain_log']/df1['hours_per_week']+1
```

```python
print(df1[['age','age_group','capital_gain','capital_gain_log','hours_per_week','income_per_hour']])
```

### 4. Feature Selection:

```python
from sklearn.ensemble import IsolationForest
```

```python
iso_forest=IsolationForest(contamination=0.05,random_state=42)
```

```python
outliers=iso_forest.fit_predict(numerical_cols)
```

```python
df1['outliers']=outliers
```

```python
df1_cleaned=df1[df1['outliers']==1].drop(columns=['outliers'])
```

```python
print("Original data shape:",df1.shape)
print("Data Shape after Outlier Removal:",df1_cleaned.shape)
```
