# Interview Questions 

#### Can you explain the difference between user-based and item-based collaborative filtering?
##### Focus: User-based looks at similar users, while item-based looks at similar items.
##### Performance: Item-based is often more scalable and works better for large, sparse datasets.
##### Recommendation Basis: User-based leverages the "people like you" approach, whereas item-based leverages "items like this" for recommendations.

#### What is collaborative filtering, and how does it work?
##### Collaborative filtering is a popular technique in recommendation systems that suggests items (such as movies, products, or songs) to users based on the preferences of other users or on item similarities. It’s based on the assumption that if two users have similar preferences, or if items have similar ratings, they can help make predictions about what someone might like.
